# Requests: HTTP for Humans

HTTP (HyperText Transfer Protocol) is the foundation of data communication on the web. It defines how messages are formatted and transmitted, and how web servers and browsers should respond to various commands.

The requests library in Python makes it easy to send HTTP requests and receive responses. It abstracts the complexities of making HTTP requests behind a simple API, making it one of the most popular libraries for web scraping and API interaction.

HTTP Methods:

| HTTP Method | Purpose                   | Example                             |
|-------------|---------------------------|-------------------------------------|
| **GET**     | Retrieve data             | Fetch a webpage                     |
| **POST**    | Submit data               | Submit a form                       |
| **PUT**     | Update or create resource | Update user profile                 |
| **DELETE**  | Remove a resource         | Delete a user account               |
| **PATCH**   | Partial modification      | Update specific user profile fields |
| **HEAD**    | Retrieve headers          | Check resource metadata             |
| **OPTIONS** | Describe options          | Determine allowed methods           |
| **TRACE**   | Message loop-back test    | Debug request path                  |



### GET Request

### POST Requests

A POST request is used to send data to a server to create/update a resource.

```python
import requests

data = {'key': 'value'}
response = requests.post('https://httpbin.org/post', data=data)
print(response.status_code)
print(response.text)
```

```python
import requests
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Fetch the JSON data
response = requests.get('http://api.open-notify.org/iss-now.json')
data = response.json()

# Extract latitude and longitude from the JSON data
longitude = float(data['iss_position']['longitude'])
latitude = float(data['iss_position']['latitude'])

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

# Create a Basemap instance
m = Basemap(projection='mill', llcrnrlat=-90, urcrnrlat=90,
            llcrnrlon=-180, urcrnrlon=180, resolution='c', ax=ax)

# Draw coastlines and countries
m.drawcoastlines()
m.drawcountries()

# Plot the ISS position
x, y = m(longitude, latitude)
m.plot(x, y, 'ro', markersize=10, label='ISS')

# Add a legend
plt.legend(loc='lower left')

# Add a title
plt.title('Current Location of the ISS')

# Show the plot
plt.show()
```
