# Poker test (test of randomness)

Create a simple "poker test" for any sequence of "0" and "1". The test should count and display the number of runs (sequences of same character).

Demonstrate your code on two following testcases. Your code should work for any other testcase possible. The output table should has the correct number of rows (according to any testcase).

The input should be inserted as the `DATA` variable containing string (exactly as in the test cases).

More information on topic of randomness tests can be foud at [wiki](https://en.wikipedia.org/wiki/Randomness_tests).


## What is the run?

Run is is a sequence of the same characters. Examples:

- run of a lenght one: 1, or 0
- run of a lenght two: 11, or 00
- run of a lenght three: 111, or 000
- etc.

Note: the only valid apperance of the ones run of a lenght three is `01110`. The `1111` does not contain runs of lenght 2, and 3. It is classified purely as a run of lenght 4.

More examples

- `111000` - contains 2 runs: 111, 000
- `011001110` - contains 5 runs: 0, 0, 00, 11, 111

## Why we need to count the runs?

Because there are prescribed ranges (intervals) how many runs we need in a sequence of ones and zeros to consider this sequence as possibly random. More reading about this standards is for example [here](https://en.wikipedia.org/wiki/FIPS_140-2)
