  <li>Discrete state space representation:
      $$\boldsymbol{x}(k+1) = \boldsymbol{Ax}(k) + \boldsymbol{Bu}(k)$$<br>
      $$\boldsymbol{y}(k) = \boldsymbol{Cx}(k) + \boldsymbol{Du}(k)$$</li> 
  <li>Make a function or class of LTI system simulator. <br>
    Input is: matrices $\boldsymbol{A, B, C, D}$, the initial condition vector $\boldsymbol{x}(0)$, time vector and input (control) vector $\boldsymbol{u}(k)$. <br>
    Output is: State variables vector $\boldsymbol{x}(k)$, output vector $\boldsymbol{y}(k)$ and time.
 
</li>
  <li>Choose any LTI system you want (at least second order system) and make a simulation using your class or function.
  </li>
  <li>Make a visualization of state variables $\boldsymbol{x}(k)$ dependent on each other <br>

    </li>
